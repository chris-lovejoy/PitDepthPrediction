## Looking at AMPLITUDE at different frequencies for the same fault, across different depths and shapes of fault

```python
AP01 = data_features.loc[(data_features['Tube_Alias'] == 'AP01') & 
                                                    (data_features['Angle'] == 0)]

waveforms_diff_feats=AP01[['Tube_Alias','Flaw_ID']+Features.DIFFERENTIAL]
waveforms_abs_feats=AP01[['Tube_Alias','Flaw_ID']+Features.ABSOLUTE]
```

```python
freq_range = [1760,1540,1320,1100,990,880,660,440,220,100] #in kHz

fig, axes = plt.subplots(3,3, figsize=(25,12))

for flaw, ax in zip(waveforms_diff_feats['Flaw_ID'].unique(), axes.flatten()):
    
    diff_feats = waveforms_diff_feats.loc[waveforms_diff_feats['Flaw_ID'] == flaw]
    abs_feats = waveforms_abs_feats.loc[waveforms_abs_feats['Flaw_ID'] == flaw]
                                       
    diff_amps = diff_feats.loc[:, diff_feats.columns.str.contains('Amp')].T
    abs_amps = abs_feats.loc[:, abs_feats.columns.str.contains('Amp')].T
    
    ax.plot(freq_range, diff_amps, label='Differential')
    ax.plot(freq_range, abs_amps, label='Absolute')
    
    ax.set(xlabel='frequencies (kHz)', ylabel='amplitudes',
       title=f'Amplitudes at different frequencies (Flaw Type {flaw})') 
    # And add variable function to give details
    
    ax.grid()
    ax.legend()
    
plt.tight_layout()
plt.show()
```

## Looking at PHASE at different frequencies for the same fault, across different depths and shapes of fault

```python
fig, axes = plt.subplots(3,3, figsize=(25,12))

for flaw, ax in zip(waveforms_diff_feats['Flaw_ID'].unique(), axes.flatten()):
    
    diff_feats = waveforms_diff_feats.loc[waveforms_diff_feats['Flaw_ID'] == flaw]
    abs_feats = waveforms_abs_feats.loc[waveforms_abs_feats['Flaw_ID'] == flaw]
                                       
    diff_amps = diff_feats.loc[:, diff_feats.columns.str.contains('Phase')].T
    abs_amps = abs_feats.loc[:, abs_feats.columns.str.contains('Phase')].T
    
    ax.plot(freq_range, diff_amps, label='Differential')
    ax.plot(freq_range, abs_amps, label='Absolute')
    
    ax.set(xlabel='frequencies (kHz)', ylabel='phases',
       title=f'Amplitudes at different frequencies (Flaw Type {flaw})') 
    # And add variable function to give details
    
    ax.grid()
    ax.legend()
    
plt.tight_layout()
plt.show()
```

CONCLUSION: phase does not strongly vary with frequency.
